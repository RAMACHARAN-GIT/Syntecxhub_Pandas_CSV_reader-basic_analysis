## Read CSV & Excel into DataFrame ##

```python
import pandas as pd
c_data = pd.read_csv("/content/student__marks.csv")
x_data = pd.read_excel("/content/student_marks.xlsx")
print(c_data.head(2)) # Top 2 rows
print("\n")
print(x_data.head(2))
print(c_data.tail(2)) # Bottom 2 rows
print(c_data.dtypes) # Data Types
print(c_data.shape) # Data Shape
```

## Compute Summary Statistics ##

```python
print(" Minimum:",c_data.min(),
      " Maximum:",c_data.max(),
      " Mean:",c_data.mean(),
      " Median:",c_data.median(),
      " Mode:",c_data.mode(),
      " Standard Deviation:",c_data.std(),
      " Variance:",c_data.var(),
      " Count:",c_data.count(),
      " Describe:",c_data.describe())
```

## Filter rows, select columns and slice subsets ##

```python
Marks_50 = c_data[c_data["Maths"] > 50]
print(Marks_50) # shows who get more than 50 marks in Maths
print(c_data[["English","Hindi"]].head(2)) # chossing two columns and top 2 rows
print("Choosing rows:\n",c_data.iloc[0:3]) # choosing top 3 rows using Index locking
print("rows and columns:\n",c_data.iloc[0:3, 2:4]) # choosing top 3 rows and 2 subjects (maths, science)
```

## Save filtered results to CSV/excel ##

```python
top_students = c_data[c_data["Maths"] > 95] # saving who got above 95 marks in maths
print(top_students)
top = top_students.to_csv("top_students.csv",index=False) # Saving as csv file
x_top = top_students.to_excel("top_students.xlsx", index=False) # Saving as excel file
```
